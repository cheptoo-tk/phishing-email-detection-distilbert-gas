```python
#import numpy and pandas for mathematical computation and data manipulation respectively
import numpy as np
import pandas as pd
#import drive package to connect this colab file with the drive where the data will be retrived from
from google.colab import drive
```

```python
from sklearn.model_selection import train_test_split
from datasets import Dataset, DatasetDict
```

**Step 3: Import the two datasets**

```python
#mount google drive to access the dataset directly from the drive
drive.mount('/content/drive')
```

```python
#Import the Email Dataset
email_df = pd.read_csv('/content/drive/MyDrive/CNS Project II/Phishing Email Detection/PED_EmailDataset.csv')

email_df.head()
```

```python
#Import the URL Dataset
url_df = pd.read_csv('/content/drive/MyDrive/CNS Project II/Phishing Email Detection/PED_URLDataset.csv')

url_df.head()
```

**Step 4: Combine the Datasets**

```python
# Step 4.1: Rename columns in both dataframes
email_df = email_df.rename(columns={'emails': 'content', 'labels': 'label'})
email_df.head()
```

```python
url_df = url_df.rename(columns={'urls': 'content', 'labels': 'label'})

url_df.head()
```

```python
# Step 4.2: Concatenate the dataframes vertically
merged_df = pd.concat([email_df, url_df], axis=0, ignore_index=True)

merged_df.head()
```

```python
# Step 4.3: Reset the index
merged_df = merged_df.reset_index(drop=True)
```

```python
# Step 4.4: Display the result
print(merged_df.head(10))
print(merged_df.columns)
print(merged_df.shape)
```

```python
# Shuffle the entire dataset
merged_df = merged_df.sample(frac=1, random_state=42).reset_index(drop=True)

merged_df.info()
```

```python
# Split off the test set (30% of the total)
train_val_df, test_df = train_test_split(merged_df, test_size=0.3, stratify=merged_df['label'], random_state=42)
```

```python
# Split the remaining data into train and validation sets
# 0.1429 test_size here will give us 10% of the total dataset (1/7 of the remaining 70%)
train_df, val_df = train_test_split(train_val_df, test_size=1/7, stratify=train_val_df['label'], random_state=42)
```

```python
#Divide the dataset into train, validation and test categories
train_val_df, test_df = train_test_split(merged_df, test_size=0.30, random_state=42)
train_df, val_df = train_test_split(train_val_df, test_size=1/7, random_state=42)
```

**Step 5: Save Merged Dataset to Google Drive**

```python
merged_df.to_csv('/content/drive/MyDrive/CNS Project II/Phishing Email Detection/PED_Datasetv2.0.csv', index=False)
print("Saved Merged dataset")
```

**Step 6: Upload Phishing Email Dataset on HuggingFacce**

```python
# Import the required libraries and log in to your Hugging Face account
from datasets import Dataset
from huggingface_hub import login

# Log in to your Hugging Face account
login()
```

```python
# Convert the pandas DataFrames to Hugging Face Datasets
train_dataset = Dataset.from_pandas(train_df)
eval_dataset = Dataset.from_pandas(val_df)
test_dataset = Dataset.from_pandas(test_df)
```

```python
#remove unnecessary column
test_dataset=test_dataset.remove_columns("__index_level_0__")
eval_dataset=eval_dataset.remove_columns("__index_level_0__")
train_dataset=train_dataset.remove_columns("__index_level_0__")
```

```python
# Create a DatasetDict
dataset_dict = DatasetDict({
    'train': train_dataset,
    'validation': eval_dataset,
    'test': test_dataset
})
```

```python
# Print some information about the splits
print("Train set size:", len(train_dataset))
print("Validation set size:", len(eval_dataset))
print("Test set size:", len(test_dataset))

print("\nLabel distribution in train set:")
print(train_df['label'].value_counts(normalize=True))
print("\nLabel distribution in validation set:")
print(val_df['label'].value_counts(normalize=True))
print("\nLabel distribution in test set:")
print(test_df['label'].value_counts(normalize=True))
```

```python
# Push the dataset to the Hugging Face Hub
dataset_name = "cybersectony/PhishingEmailDetectionv2.0"

dataset_dict.push_to_hub(dataset_name)
```
